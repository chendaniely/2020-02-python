```python
df1 = pd.DataFrame({
    'a': ['a1', 'a2', 'a3', 'a4', 'a5'],
    'b': [1, 1, 2, 2, 3]
})
df1
```

```python
df2 = pd.DataFrame({
    'a2': ['2a1', '2a2', '2a3', '2a4', '2a5'],
    'b2': [1, 1, 1, 2, 2]
})
df2
```

```python
m2m = pd.merge(df1, df2, left_on='b', right_on='b2')
m2m
```

```python
assert len(df1) == len(df2)
```

```python
assert len(df1) == len(m2m)
```
