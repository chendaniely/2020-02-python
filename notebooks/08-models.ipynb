```python
import pandas as pd
import seaborn as sns
from sklearn import linear_model
import statsmodels.formula.api as smf
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
model = smf.ols("tip ~ total_bill", data=tips)
```

```python
results = model.fit()
```

```python
results.summary()
```

```python
model = smf.ols("tip ~ total_bill + sex + smoker + sex", data=tips).fit()
```

```python
model.summary()
```

```python
lr = linear_model.LinearRegression()
```

```python
# lr.fit(X=tips['total_bill'].values.reshape(1, -1), y=tips['tip'])
```

```python
lr.fit(X=tips[['total_bill', 'size']], y=tips['tip'])
```

```python
lr.coef_
```

```python
lr.intercept_
```
