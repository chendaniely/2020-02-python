```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, fit_reg=False)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', hue='time', data=tips, fit_reg=False)
plt.show()
```

```python
tips.info()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, fit_reg=False,
          hue='sex', col='day', row='smoker')
plt.show()
```
